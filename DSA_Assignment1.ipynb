# Histogram representation of SL

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10, 7))
x = data["SL"]
  
plt.hist(x, bins = 20, color = "green", rwidth=0.9)
plt.title("SL in cm")
plt.xlabel("SL in cm")
plt.ylabel("Count")
```

# Box Plot for Iris Data

```python
new_data = data[["SL", "SW", "PL", "PW"]]
plt.figure(figsize = (10, 7))
new_data.boxplot()
```

# Plotting using Seaborn on iris dataset

```python
import seaborn as sns
iris=pd.read_excel("D:\python\DSA-assignment1\iris.xls")
iris = sns.load_dataset('iris')
sns.pairplot(iris)
```

# Violin Plot

```python
sns.violinplot(x="species", y="petal_length", data=iris, size=6)
```

# Heatmap

```python
plt.figure(figsize=(7,4))
sns.heatmap(iris.corr(),annot=True)
```
